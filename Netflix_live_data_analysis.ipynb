### Q10. In which year highest number of the TV Shows & Movies were released ? Show with Bar Graph.

### pd.to_datetime() method.........

Here we're going to make a new column additional by changing the datatypes of Release_Date column

```python
df['Date_Time']=pd.to_datetime(df['Release_Date'],format='mixed')

#Note: here we used 'mixed' in the format, why?
#Ans-> coz, the column that datatypes we wanted to changed had data& time written differently sometimes date was in the beginning or sometimes time was in the beginning...
```
